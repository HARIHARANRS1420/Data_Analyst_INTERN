```python
# Convert df2 to timestamp
df2['date_converted'] = pd.to_datetime(df2['Timestamp IST'], dayfirst=True)
df2 = df2.set_index('date_converted')
df2
```

```python
# Resample df1 to daily frequency and aggregate
df1_daily = df1.resample('D').agg({
    'value': 'mean',
    'classification': lambda x: x.mode()[0] if not x.empty else None
})
df1_daily.head()
```

```python
# Resample df2 to daily frequency and aggregate
df2_daily = df2.resample('D').agg({
    'Execution Price': 'mean',
    'Size Tokens': 'sum',
    'Size USD': 'sum',
    'Start Position': 'mean',
    'Closed PnL': 'sum',
    'Fee': 'sum'
})
df2_daily.head()
```

```python
# Merge the two daily dataframes
merged_df = pd.merge(df1_daily, df2_daily, left_index=True, right_index=True, how='outer')
merged_df
```

```python
# Part B
# Calculate max drawdown for Fear and Greed days
def calculate_max_drawdown(pnl_series):
    if pnl_series.empty:
        return np.nan

    cumulative_pnl = pnl_series.cumsum()
    peak = cumulative_pnl.expanding(min_periods=1).max()
    drawdown = (cumulative_pnl - peak) / peak
    max_drawdown = drawdown.min() * 100
    return max_drawdown

fear_days_pnl = merged_df[merged_df['classification'] == 'Fear']['Closed PnL'].dropna()
greed_days_pnl = merged_df[merged_df['classification'] == 'Greed']['Closed PnL'].dropna()

max_drawdown_fear = calculate_max_drawdown(fear_days_pnl)
max_drawdown_greed = calculate_max_drawdown(greed_days_pnl)

print(f"Maximum Drawdown on Fear Days: {max_drawdown_fear:.2f}%")
print(f"Maximum Drawdown on Greed Days: {max_drawdown_greed:.2f}%")
```

```python
# Calculate average leverage per account
average_leverage_per_account = df2_leverage.groupby('Account')['leverage'].mean().reset_index()
average_leverage_per_account = average_leverage_per_account.rename(columns={'leverage': 'Average_Leverage'})

# Calculate total trading days per account
trading_frequency_per_account = df2.groupby('Account').apply(lambda x: x.index.normalize().nunique()).reset_index(name='Trading_Days')

# Calculate total PnL per account
total_pnl_per_account = df2.groupby('Account')['Closed PnL'].sum().reset_index()
total_pnl_per_account = total_pnl_per_account.rename(columns={'Closed PnL': 'Total_PnL'})

# Merge all metrics into a single DataFrame
trader_profiles = average_leverage_per_account.merge(trading_frequency_per_account, on='Account', how='left')
trader_profiles = trader_profiles.merge(total_pnl_per_account, on='Account', how='left')

trader_profiles.head()
```

```python
#Segmentation criteria based on medians
median_leverage = trader_profiles['Average_Leverage'].median()
median_trading_days = trader_profiles['Trading_Days'].median()
median_pnl = trader_profiles['Total_PnL'].median()

#Segment traders
trader_profiles['Leverage_Segment'] = trader_profiles['Average_Leverage'].apply(lambda x: 'High Leverage User' if x > median_leverage else 'Low Leverage User')
trader_profiles['Frequency_Segment'] = trader_profiles['Trading_Days'].apply(lambda x: 'Frequent Trader' if x > median_trading_days else 'Infrequent Trader')
trader_profiles['PnL_Segment'] = trader_profiles['Total_PnL'].apply(lambda x: 'Consistent Winner' if x > median_pnl else 'Inconsistent Trader')

# Trader counts by Leverage Segment
display(trader_profiles['Leverage_Segment'].value_counts())

# Trader counts by Frequency Segment
display(trader_profiles['Frequency_Segment'].value_counts())

# Trader counts by PnL Segment
display(trader_profiles['PnL_Segment'].value_counts())

# Average PnL by Segment Combinations
display(trader_profiles.groupby(['Leverage_Segment', 'Frequency_Segment', 'PnL_Segment'])['Total_PnL'].mean().unstack())
```

```python
# Prepare data for plotting
avg_pnl_by_segment = trader_profiles.groupby(['Leverage_Segment', 'Frequency_Segment', 'PnL_Segment'])['Total_PnL'].mean().reset_index()
avg_pnl_by_segment['Segment_Combination'] = avg_pnl_by_segment['Leverage_Segment'] + ' - ' + avg_pnl_by_segment['Frequency_Segment'] + ' - ' + avg_pnl_by_segment['PnL_Segment']

# Create the bar chart
plt.figure(figsize=(14, 8))
sns.barplot(x='Segment_Combination', y='Total_PnL', hue='PnL_Segment', data=avg_pnl_by_segment, palette='viridis')
plt.title('Average PnL Across Trader Segment Combinations')
plt.xlabel('Trader Segment Combination')
plt.ylabel('Average Total PnL')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Conclusion
Based on my analysis of traders, looking at Total PnL, leverage, and trading frequency, here are two simple strategy ideas:

For Successful Traders (Regular Profit, Less Trading, Low Leverage):
If you are making profits consistently by trading less often and using low leverage, continue this method. It shows that taking only good trading opportunities and managing risk carefully can give better results.

For Traders with Losses or Unstable Results (High Leverage, Frequent Trading):
If your results are not consistent and you trade often with high leverage, try to reduce both leverage and the number of trades. This can help avoid big losses and make your profits more stable, especially when the market is highly volatile or fearful.
